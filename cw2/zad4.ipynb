```python
import scipy.stats as scs
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(16, 6))

ax1 = fig.add_subplot(131)
p = 0.3
x1 = [0, 1]
rv1 = scs.bernoulli(p)
ax1.vlines(x1, 0, rv1.pmf(x1), colors='b', linestyles='-', lw=1, label='Bernoulli')
ax1.legend(loc='best', frameon=False)

ax2 = fig.add_subplot(132)
n = 1
x2 = [0, 1]
rv2 = scs.binom(n, p)
ax2.vlines(x2, 0, rv2.pmf(x2), colors='b', linestyles='-', lw=1, label='Binomial')
ax2.legend(loc='best', frameon=False)

ax3 = fig.add_subplot(133)
mu = 2
x3 = np.arange(scs.poisson.ppf(0.01, mu), scs.poisson.ppf(0.99, mu))
rv3 = scs.poisson(mu)
ax3.vlines(x3, 0, rv3.pmf(x3), colors='b', linestyles='-', lw=1, label='Poisson')
ax3.legend(loc='best', frameon=False)
plt.show()
```
