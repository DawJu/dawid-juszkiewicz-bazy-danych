```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(16, 6))

ax1 = fig.add_subplot(131)
data1 = norm.rvs(loc=1, scale=2, size=100)
ax1.hist(data1, histtype='bar', label='1; 2')
ax1.legend(loc='best', frameon=False)

ax2 = fig.add_subplot(132)
data2 = norm.rvs(loc=0, scale=1, size=100)
ax2.hist(data2, histtype='bar', label='0; 1')
ax2.legend(loc='best', frameon=False)

ax3 = fig.add_subplot(133)
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
data3 = norm(loc=-1, scale=0.5)
ax3.plot(x, data3.pdf(x), 'b', lw=3, label='-1; 0.5')
ax3.legend(loc='best')
plt.show()
```
