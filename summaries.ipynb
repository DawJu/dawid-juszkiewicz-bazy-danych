```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/daily_data.csv')
df['date'] = pd.to_datetime(df['date'])
df_war = df[df['location'] == 'Warszawa']
df_war.head()
```

```python
df.dtypes
```

```python
df_dec = df_war[df_war['date'].dt.month == 12]
df_dec.head()
```

```python
plt.plot(df_dec['date'], df_dec['temperature_min'], label='min temp')
plt.plot(df_dec['date'], df_dec['temperature_max'], label='max temp')
plt.xlabel('date')
plt.ylabel('temperature (degrees Celsius)')
plt.title('Lowest and highest temperature in Warsaw, December 2022')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(df_dec['date'], df_dec['rain_sum'])
plt.xlabel('date')
plt.ylabel('rain (mm)')
plt.title('Rain in Warsaw, December 2022')
plt.xticks(rotation=45)
plt.show()
```
